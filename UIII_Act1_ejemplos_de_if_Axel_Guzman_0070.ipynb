#### Ejemplo de if

```python
# 1.Ejemplo de if simple con operadores relacionales
# Este tipo de estructura ejecuta un bloque de codigo si la condicion se cumple

edad = 18 # Asigna el valor 18 a la variable 'edad'
if edad >= 18: # Evalúa si 'edad' es mayor o igual a 18 (operador relacional)
    print("Eres mayor de edad.") # Se ejecuta solo si la condición anterior es True
```

```python
# 2. Ejemplo de if else con operadores logicos relacionales
# se utiliza para elegir entre dos opciones exclusivas

numero = 10 # Define un número entero
if numero % 2 == 0: # Usa el operador aritmético módulo (%) y el relacional (==)
    print("El número es par.") # Se ejecuta si el resto de dividir por 2 es 0
else: # Bloque opcional para cuando la condición del 'if' es False
    print("El número es impar.") # Se ejecuta si la condición fue False
```

```python
# 3. Ejemplo de if elif else (Uso de operadores lógicos, relacionales y aritméticos)
# Permite evaluar múltiples condiciones en orden sucesivo

nota = 85 # Variable con una calificación
asistencia = 90 # Variable con porcentaje de asistencia

if nota >= 90 and asistencia >= 80: # Evalúa dos condiciones con el operador lógico 'and'
    print("Excelente desempeño.") # Ejecuta si ambas condiciones son True
elif nota >= 70 or asistencia >= 95: # Evalúa condiciones alternativas con 'or'
    print("Aprobado por notas o gran asistencia.") # Ejecuta si al menos una es True
else: # Caso por defecto
    print("Debes esforzarte más.") # Ejecuta si ninguna de las anteriores se cumplió
```

```python
# 4. Ejemplo combinado (Cálculo de descuento)
# Este ejemplo muestra cómo se integran operaciones aritméticas complejas dentro de la lógica de decisión.

precio_unitario = 100 # Precio base
cantidad = 12 # Cantidad comprada

if cantidad > 10: # Comprueba si hay más de 10 artículos
    total = (precio_unitario * cantidad) * 0.90 # Aplica 10% de descuento (aritmética)
    print(f"Total con descuento: {total}") # Muestra el resultado calculado
else:
    total = precio_unitario * cantidad # Calcula el total sin descuento
    print(f"Total sin descuento: {total}") # Muestra el resultado
```
